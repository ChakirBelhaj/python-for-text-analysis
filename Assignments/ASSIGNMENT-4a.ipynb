````python
#https://stackoverflow.com/questions/59811578/how-to-read-tsv-file-without-packages
#https://stackoverflow.com/questions/25896698/create-a-dictionary-from-arrays-of-keys-and-values/25896699
def read_csv(input_file, delimiter=","):
    input = open(input_file,"r")
    rawText = input.read()
    lines = rawText.split("\n")
    listOfLists = []    
    keys = []
    for index, k in enumerate(lines) : 
        if(index == 0):
            keys = k.split(delimiter)
        else:
            values = k.split(delimiter)
            listOfLists.append(dict(zip(keys, values)))
            
    return listOfLists


# test your function here
filename = "../Data/csv_data/trump_facebook.tsv"
status_updates = read_csv(filename, delimiter="\t") 
status_updates[0:2]
````

Define a function called **`get_update_most_responded_to()`** that has the following parameters:

* **`response_type`** (keyword parameter with default string `"likes"`) 

The fuction should find the status update that received the highest number of possible reactions to a Facebook status (emoji such as 'angrys', 'comments', 'hahas', etc. - anything that starts with 'num_'). It should return three strings: the **`status_message`**, the **`status_type`** and the **`status_link`** of this particular status update.


````python
def get_update_most_responded_to(status_updates, response_type):
    best_response = status_updates[0]

    for item in status_updates:
        if int(item["num_" + response_type]) > int(best_response["num_" + response_type]):
            best_response = item
    return [best_response['status_message'], best_response['status_type'], best_response['status_link']]

get_update_most_responded_to(status_updates,"sads")
# test your function here
````
